```python
movies = df[['title', 'overview', 'genres', 'keywords', 'cast', 'director']].copy() #store only required attributes
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
movies.shape
```

```python
import ast #convert stringified lists -> python lists
```

```python
def convert(text):
    if isinstance(text, list): #if already list pass through
        return [i['name'] for i in text]
    try:
        return [i['name'] for i in ast.literal_eval(text)] #if string convert to list
    except:
        return []
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies['director'] = movies['director'].apply(lambda x: [x]) #convert to list
```

```python
#Remove spaces so names don’t split:
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['director'] = movies['director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#tags column has all details in sequence
movies['tags'] = (
    movies['overview'] + " " +
    movies['genres'].apply(lambda x:" ".join(x)) + " " +
    movies['keywords'].apply(lambda x:" ".join(x)) + " " +
    movies['cast'].apply(lambda x:" ".join(x)) + " " +
    movies['director'].apply(lambda x:" ".join(x))
)
```

```python
final_df = movies[['title', 'tags']]
final_df['tags'] = final_df['tags'].apply(lambda x: x.lower())
final_df.head()
# Now each movie = one text document
```

```python
# from sklearn.feature_extraction.text import CountVectorizer #converts words to vectors

# cv = CountVectorizer(max_features=5000, stop_words='english')
# vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

vectors = tfidf.fit_transform(final_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#apply cosine similarity to get similar movies
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie = movie.lower()
    if movie not in final_df['title'].str.lower().values:
        print("❌ Movie not found. Please check the title.")
        return

    index = final_df[final_df['title'].str.lower() == movie].index[0] #match input to movie in dataset and store info
    distances = similarity[index] #match similarity b/w info of movies

    movies_list = sorted(
        list(enumerate(distances)), #create a list of distances in form of (movie_index, similarity_score)
        reverse=True, #highest first
        key=lambda x: x[1] #sort based on similarity_score
    )[1:6] #select top 5 leaving the movie itself which has score 1.0

    print("Recommended movies:\n")
    for i in movies_list:
        print(final_df.iloc[i[0]].title)
```

```python
recommend("Avatar")
```
